```python
#import library
import pandas as pd   #For data manipulation and analysis
```

```python
#Load the Dataset
data = pd.read_csv("Downloads/file.csv")
```

```python
data.shape  #return dimensions of dataset
```

```python
data.head(20)  #return top 5 rows of dataset
```

```python
#Check if there is duplicated data
data.duplicated()   #returns a series of boolean, true if previous entry is identical.
```

```python
data.duplicated().sum()
```

```python
data.loc[data.duplicated(), :] #return duplicate rows
```

```python
 data.loc[data.duplicated(keep=False), :]  # return all identical rows
```

```python
data.drop_duplicates(keep='first') # the keep parameter is set to First, so that only First Unique values are taken and the duplicate values are removed
```

```python
data.drop_duplicates(keep='first').shape
```

```python
data.drop_duplicates(keep=False) #the keep parameter is set to False, so that only Unique values are taken and all the identical values are removed
```

```python
data.drop_duplicates(keep=False).shape
```
